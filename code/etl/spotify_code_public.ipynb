```python
# Calculate the timestamp for one month ago and now
end_date = datetime.datetime.now()
start_date = end_date - datetime.timedelta(days=30)

# Convert to milliseconds for Spotify API
start_timestamp = int(start_date.timestamp()) * 1000

# Initialize an empty list to accumulate all songs
all_songs = []

# Iterate over each day in the last month
for day in range(30):
    # Calculate the start timestamp for the current day
    after_timestamp = start_timestamp + (day * 86400000)  # Start of the day

    # Fetch recently played tracks for the current day
    results = sp.current_user_recently_played(limit=50, after=after_timestamp)
    for item in results['items']:
        track_name = item['track']['name']
        artist_name = item['track']['artists'][0]['name']
        track_id = item['track']['id']
        played_date = item['played_at']
        # Append each song as a tuple to the all_songs list
        all_songs.append((track_name, artist_name, track_id, played_date))

# After the loop, print the total number of songs collected
print(f"Total songs collected: {len(all_songs)}")
```

```python
print(all_songs)
```

```python
# Create a pandas DataFrame from the all_songs list
df = pd.DataFrame(all_songs, columns=['Track Name', 'Artist Name', 'Track ID', 'Day_Played'])
#Extract just the day from the 'played_at' column
df['Day_Played'] = pd.to_datetime(df['Day_Played']).dt.date


#get the audio features for each track 
audio_features = []
for i in range(0, len(df), 50):
    track_ids = df['Track ID'][i:i+50]
    features = sp.audio_features(track_ids)
    for feature in features:
        audio_features.append(feature)

#add the audio features to the dataframe
df_audio_features = pd.DataFrame(audio_features)
df = pd.concat([df, df_audio_features], axis=1)
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('spotify_data.csv', index=False)
```

```python
# example code for pulling all of Taylor Swifts albums
taylor_uri = 'spotify:artist:06HL4z0CvFAxyc27GXpf02'

results = []

#pull the albumns and list the tracks within each of them as artist name and track id 
albums = sp.artist_albums(taylor_uri, album_type='album')
for album in albums['items']:
    album_name = album['name']
    tracks = sp.album_tracks(album['id'])
    for track in tracks['items']:
        track_name = track['name']
        track_id = track['id']
        results.append((album_name, track_name, track_id))

#convert the results into a dataframe
df_taylor = pd.DataFrame(results, columns=['Album Name', 'Track Name', 'Track ID'])
df_taylor

```

```python
#get the audio features for each track
audio_features = []
for i in range(0, len(df_taylor), 50):
    track_ids = df_taylor['Track ID'][i:i+50]
    features = sp.audio_features(track_ids)
    for feature in features:
        audio_features.append(feature)

#add the audio features to the dataframe
df_audio_features = pd.DataFrame(audio_features)
df_taylor = pd.concat([df_taylor, df_audio_features], axis=1)
df_taylor
```

```python
#save the dataframe to a csv file
df_taylor.to_csv('taylor_swift.csv', index=False)
```

```python
#Pull the top 50 songs from the "Billboard Hot 100" playlist and grab the track name, artist name, and track id
playlist_id = '37i9dQZEVXbLRQDuF5jeBp'
results = sp.playlist_tracks(playlist_id)
tracks = results['items']
billboard = []
for track in tracks:
    track_name = track['track']['name']
    artist_name = track['track']['artists'][0]['name']
    track_id = track['track']['id']
    billboard.append((track_name, artist_name, track_id))

#convert the results into a dataframe
df_billboard = pd.DataFrame(billboard, columns=['Track Name', 'Artist Name', 'Track ID'])
df_billboard

```
